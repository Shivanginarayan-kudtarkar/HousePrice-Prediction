```python
x=df.drop('House_Price',axis=1)
y=df['House_Price']
x.info()
```

```python
x.shape
x
```

```python
df
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
linear_regressor=LinearRegression()
```

```python
linear_regressor.fit(x_train,y_train)
```

```python
y_pred_lr=linear_regressor.predict(x_test)
y_pred_lr
```

```python
print("Accuracy of LR on Training Set",linear_regressor.score(x_train,y_train))
print("Accuracy of LR on Testing Set",linear_regressor.score(x_test,y_test))
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse

print("---Linear Regression--Model Evaluation---")
print("Mean absolute Error(MAE):",mae(y_test,y_pred_lr))
print("Mean squared Error(MSE):",mse(y_test,y_pred_lr))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
```

```python
# decision tree 
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
y_pred_dt=dt.predict(x_test)
y_pred_dt
```

```python
print("Accuracy of DT on Training Set",dt.score(x_train,y_train))
print("Accuracy of DT on Testing Set",dt.score(x_test,y_test))
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse

print("---Decision Tree Regression--Model Evaluation---")
print("Mean absolute Error(MAE):",mae(y_test,y_pred_dt))
print("Mean squared Error(MSE):",mse(y_test,y_pred_dt))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_dt)))
```

```python
# Random forest
from sklearn.ensemble import RandomForestRegressor
random_regressor=RandomForestRegressor()
random_regressor.fit(x_train,y_train)
```

```python
y_pred_rf=random_regressor.predict(x_test)
y_pred_rf
```

```python
print("Accuracy of RF on Training Set",random_regressor.score(x_train,y_train))
print("Accuracy of RF on Testing Set",random_regressor.score(x_test,y_test))
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse

print("---Random Forest--Model Evaluation---")
print("Mean absolute Error(MAE):",mae(y_test,y_pred_rf))
print("Mean squared Error(MSE):",mse(y_test,y_pred_rf))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_rf)))
```

```python
from sklearn.svm import SVR

svr=SVR()
svr.fit(x_train,y_train)
```

```python
svr_pred=svr.predict(x_test)
svr_pred
```

```python
print("Accuracy of SVR on Training Set",random_regressor.score(x_train,y_train))
print("Accuracy of SVR on Testing Set",random_regressor.score(x_test,y_test))
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse

print("---SVR--Model Evaluation---")
print("Mean absolute Error(MAE):",mae(y_test,svr_pred))
print("Mean squared Error(MSE):",mse(y_test,svr_pred))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,svr_pred)))
```
